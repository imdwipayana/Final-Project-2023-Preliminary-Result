```python
import pandas as pd

import warnings
warnings.filterwarnings("ignore")

pd.set_option('max_columns',500)
```

```python
data = pd.read_csv('airbnb.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.dtypes
```

```python
data.loc[ data['Guest Satisfaction'] <= 40, 'Guest Satisfaction'] = 0
data.loc[(data['Guest Satisfaction'] > 40) & (data['Guest Satisfaction'] <= 60), 'Guest Satisfaction'] = 1
data.loc[(data['Guest Satisfaction'] > 60) & (data['Guest Satisfaction'] <= 80), 'Guest Satisfaction'] = 2
data.loc[ data['Guest Satisfaction'] > 80, 'Guest Satisfaction'] = 4
```

```python
data.head()
```

```python
data['Guest Satisfaction'].unique()
```

```python
data = pd.get_dummies(data)
```

```python
X = data.drop(['Guest Satisfaction'],axis=1)
y = data['Guest Satisfaction']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_lr = LogisticRegression()
```

```python
model_lr.fit(X_train,y_train)
```

```python
model_lr.score(X_test,y_test)
```
